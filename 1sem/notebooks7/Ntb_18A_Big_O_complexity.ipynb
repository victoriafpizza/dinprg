### 4 - Function to compute the time and memory used after a command run.

#### https://docs.python.org/3/library/tracemalloc.html

```python
def monit(A):
  # A = command to be done
  # process = psutil.Process(os.getpid())
  start = time.time() # time start
  tracemalloc.start () # tracemalloc start
  # start_memory_usage = process.memory_info().rss
  A
  memo_used = EngNumber(psutil.Process().memory_info().rss)
  size,peak = tracemalloc.get_traced_memory()
  size = EngNumber(size)
  peak = EngNumber(peak)
  # stop_memory_usage = process.memory_info().rss
 #  memory_usage = (stop_memory_usage) - (start_memory_usage)
  # memory_usage = EngNumber(memory_usage)
  tracemalloc.reset_peak()
  stop = time.time()
  dif = (stop - start)
  dif = EngNumber(dif)
  print('the time (s) and memory (size,peak) used (B) were:',dif,' ','size = ',size,' ','peak =', peak)
  print('the time (s) and memory (size) used (B) were:',dif, ' ', 'size =', memo_used)
```

### Bonus: function to find the file delimiter while reading an input file.

```python
def get_delimiter(file_path, bytes = 4096):
    sniffer = csv.Sniffer()
    data = open(file_path, "r",encoding='latin-1').read(bytes)
    delimiter = sniffer.sniff(data).delimiter
    return delimiter
```
