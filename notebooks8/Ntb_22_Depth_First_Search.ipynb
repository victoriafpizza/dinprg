## Depth-First-Search Algorithm

Trying to search all the paths in a Graph can be done exploring senses till reaching a 'dead-end'. This means to search all the possibilities till to wind up in a node there is no other way to proceed. Once we have reached this stage, we use 'backtracking' (move backwards in the same way already trailed). This procedure is done in a loop till to collect all the possible paths to find the 'dead end'.

Depth-Search means this concept, trying to reach the dead end, and after that, backtracking to find another possible dead end, which can be the same one, but towards another path or sequence of nodes. it sounds more vertical, while the Breadth-Search is more 'horizontal'.

In a set of steps:
A - Start by putting any one of the graph's vertex on top of the stack.
B - Take the top item of the stack and add it to the visited list of the vertex, similar as done in the Breadth-Search.
C - Next, create a list of that adjacent node of the vertex, or nodes that have a link with the origin where we are.
D - Add the ones which aren't in the visited list of vertexes to the top of the stack.
E - Repeat steps B to D, until the stack is empty.

#### Depth-Search code

```python
graph = { '5' : ['3','7'],  '3' : ['2', '4'],
  '7' : ['8'],  '2' : [],  '4' : ['8'],  '8' : []}
```

```python
visited = set() # Set to keep track of visited nodes of graph.

def dfs(visited, graph, node):  #function for dfs
    if node not in visited:
        print (node)
        visited.add(node)
        for neighbour in graph[node]:
            dfs(visited, graph, neighbour)
```
