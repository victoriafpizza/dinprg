## Breadth_Search: another way to find paths in Graphs or Trees.

Como visto no notebook anterior (20), nem sempre se consegue obter o caminho otimizado, ou se avaliar todas as alternativas de forma otimizada.

Para isso, deve-se usar algoritmos de busca específicos para Grafos ou Árvores, assim como feito para listas/arrays, que foram o Bubble, Merge, Heap ou Quick. O Algoritmo Breadth_Search é um deles, onde 'Breadth' significa largura ou extensão horizontal. O conceito é varrer horizontalmente, tudo o que for possível, até avançar para o novo nível, de forma sequencial e recursiva.

Start: definir um Grafo, por meio de nós e arestas, usando dicinário e listas, como já comentado no notebook anterior (20).

```python
graph = { '5' : ['3','7'],  '3' : ['2', '4'], '7' : ['8'],
  '2' : [], '4' : ['8'], '8' : [] }
```

```python
visited = [] # List for visited nodes.
queue = []     #Initialize a queue. Queue = Fila.

def bfs(visited, graph, node): #function for BFS
# visited = list of nodes already visited
# graph = graph to be checked
# node = start node
  visited.append(node) # the start node is the first to be considered as 'visited'.
  queue.append(node) # see above

  while queue:          # Creating loop to visit each node
    m = queue.pop(0) # take the first
    print (m, end = " ")

    for neighbour in graph[m]:
      # neighbour is the node with link to the node considered
      if neighbour not in visited:
        visited.append(neighbour)
        queue.append(neighbour)
```

```python
# Driver Code
print("Following is the Breadth-First Search")
bfs(visited, graph, '5')    # function calling
```
