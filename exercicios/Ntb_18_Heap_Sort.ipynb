1 - A "heap" in Python is a data structure based on a unique binary tree designed to efficiently access the smallest or largest element in a collection of items. Thus, there are a 'maximum heap algorithm' and a 'minimum heap algorithm'.

2 - Binary Tree = from a 'father' node, there are only two 'children' nodes.

Suppose a binary tree with 3 elements: 3, 2, 1. The father node has the number '3', and the children node are then '2' and '1'. In the logic, the smallest shall be on the left, thus the right binary tree is: 3,1,2.

With 3 elements, the binary level is '1', because it has the two children in this first level. If an Array has 7 elements: [9,5,4,6,3,2,1], the first 'father node' is 9, followed by 5 and 4. The node '5' is followed by 6 and 3, and the node '4' is followed by 2 and 1.

Checking the binary three above, the numbers are not in the correct order. There are 'children' nodes higher than the 'father' node, and the highest 'children' node must be on the right position. However, this order can be the opposite, in terms of the 'father' node.

The heap data structure is basically used as a "heapsort algorithm" to sort the elements in an array or a list. These algorithms can be used in priority queues, order statistics etc. The main application of heap is to set an element order of an array, list, tuple etc.

3 - There are two possibilities linking the nodes: the prior node can be higher than the following nodes ('maximum heap'), or the prior node can be smaller than the following nodes ('minimum heap').

While using binary trees and heap, the smallest element on a node goes to the left and the highest to the right. Another key observation: the analysis is done by 'node', one by one, repeating the algorithm till the stage with all nodes following the logic (maximum or minimum heap).

In the maximum Heap Sort = the prior node must be higher than the two following nodes. This means the node with 'children' must be higher than the children. The tree is binary, thus there are just two children connected to the 'father' node. In the maximum Heap, the algorithm starts from the end or bottom part of the three, amd moves upwards.

In the minimum Heap Sort = the prior or 'father' node must be smaller than the two following nodes, or 'children' nodes.

4 - Remember = the smallest node shall be on the left!!

5 - Heap Algorithm - Maximum option (code)
Let´s implement the code by 'functions'.
A set of functions becomes a 'Class'.
There must be a first argument = 'list' or a'array', from where the elements will be checked and ordered.
The binary trees have levels, which is related to the lenght of the array. See the examples above.
Thus, there are an index for the level (k) and an index for the left or right position (at the same level).

```python
from ctypes import c_int16

# k = integer, index related to the father nodes.
# for k = 0, there are the two first 'children' nodes = A[1] & A[2]
# for k = 1, there are two 'children' nodes, related to A[1]: A[3] & A[4]
# for k = 2, there are two 'children' nodes, related to A[2]: A[5] & A[6]

def left(k):
    return 2 * k + 1 # the function provides an odd number (número ímpar)

def right(k):
    return 2 * k + 2

def max_heapify(A,k):
  # A = list or array with elements
  # k = index, integer - explained above - father node index
    l = left(k) # odd number, generated as above indicated
    r = right(k) # even number, see the function above
    if l < len(A) and A[l] > A[k]:
        largest = l # this means the largest value in on the left children node
    else:
        largest = k # this means the father node is higher than the left node
    if r < len(A) and A[r] > A[largest]:
        largest = r # the same concept above
    if largest != k: # for the comparison between the left and right children nodes
        A[k], A[largest] = A[largest], A[k] # this works to keep the largest on the right
        A = max_heapify(A, largest) # recursion = applying the function with the new index
    return A

# Doing so, the largest number goes from the bottom to the top
# To apply the algorithm, the function below does the job, from the bottom to the top.
def build_max_heap(A):
  # A = the list or array input
    n = int((len(A)//2)-1)
    for k in range(n, -1, -1):
  # range = built-in function, with three parameters (start, stop, step)
  # stop = '-1', means the last value will be '0'.
  # step = '-1', means decreasing by -1, at every step.
       A = max_heapify(A,k)
    return A
```

```python
B = [12,6,5,7,3,2,10]
C = build_max_heap(B)
print(C)
```
