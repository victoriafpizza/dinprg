## 1) The algorithm Bubble Sort deals with the highest value to the right.
For this purpose, we work with the index of the element within a list or array.
The simplest comparison envolver two elements, one closer to the other by just one position index.

```python
def bubble_sort(array):
    # the instance or argument here is an array = set of elements where the position inside matters
    n = len(array)
    # n = is a integer number related to the quantity of elements of an array
    for i in range(n-1): # associated index to the number of elements inside the array
        for j in range(0, n-1):
          # j is the index associated to the specific position within the array
          # could be n-i-1 as, e.g., top value is guaranteed to have bubbled up in 1st iteration
            if array[j] > array[j+1]:
                array[j], array[j+1] = array[j+1], array[j] # this is the new order, shifted between just two close elements
          # make attention there is no need to state the other option else. There are two possibilities in the end.

    return array
    # the output is a new array, after all the comparison and changes.
```

## 2) The algorithm Merge Sort works first splitting and then sorting.
First break the array into smaller parts. Sort theses smaller parts. After that, set the sorted parts together and do the sort again.
Repeat this approach till reaching the full array.

```python
def merge_sort(my_list):
  # using list instead of array, so we can .pop() (np arrays don't pop with built-in method)
  # .pop() will be used in the splitting stage

    if len(my_list) > 1:
      # remember to treat the simplest case first
      # if length is 1, no need to sort (at deepest recursion, some will have len 1, others len 2)

        # 1. Halve:
        # Halve = Ã© um verbo, que significa dividir na metade
        # Half = metade
        midpoint = len(my_list) // 2 # exact division, it will be a single number to be associated to the list element index.
        left_half = my_list[:midpoint] # check the midpoint is a index now and `left_half` is all the elements on the left till the midpoint
        right_half = my_list[midpoint:] # now is the opposite, we need to select the right part.

        # 2. Sort (recursively):
        # Sort = to put in an order
        # See how we use `recursion`to to set a specific order
        left_half = merge_sort(left_half)
        right_half = merge_sort(right_half)

        my_list = []
        # this is an empty lits, it will receive the two sorted parts (left and right), or the merge phase

        # 3. Merge:
        while len(left_half)>0 and len(right_half)>0:
          # Note we are handling the two halfs in the same command. This command deals with the simplest cases, it must use a AND association.

            if left_half[0] < right_half[0]:
              # comparing the two halfs, with the same element index
                my_list.append(left_half.pop(0))
                # pop first element
                # pop is useful here to detach the element, not to scrap or erase it
                # the smallest element is set into the list, thus the smaller ones will be kept to the left.

            else:
              # 1st element of right half < 1st element of left
              # here we need the else, because it handles the other possibility - the right
                my_list.append(right_half.pop(0))

        # If any elements remain from either half, they must be the largest value:
        for i in left_half:
            my_list.append(i)
        for i in right_half:
            my_list.append(i)

    return my_list
    # the output is a new list, with an asceding order from the left to the right
```

## 3) Quick Sort - make comparisons and sort with a pivot
Similar to the Merge Sort and uses Bubble Sort somehow.
The reference element is a `pivot`, chosen by any criterion, such as the last one, the middle, the median etc...
This algorithm has three separate sections as follow:

*   Partition
*   Sorting
*   Algorithm application - junction of the two above



```python
# First Part - partition
# to split the set of elements
def partition(array, low, high):
    # this function will give a index number
    # 1. Use the final element as the pivot:
    pivot = array[high]
    # pivot = reference index-element

    i = (low-1) # index of element on the small side
    for j in range(low, high):
      # j = index to scroll or walk through
      # like a bubble swap, we compare pivot to all elements between low and high
        if array[j] < pivot:
          # if value < pivot, it's correctly on left so increment the smaller element's index...
            i = i+1
            array[i], array[j] = array[j], array[i]
            # like Bubble Sort = smallest to the left, largest to the right
    array[i+1], array[high] = array[high], array[i+1]
    # ...otherwise, swap greater value to right of pivot (now what we want)
    return(i+1) # return index of correctly positioned pivot
```

```python
# Second part - pivot action
def _quick_sort(array, low, high):
  # array = the set of elements to be sorted
  # low = a reference index number
  # high = another reference index number
  # low shall be smaller than high

    if low < high: # when low reaches high, we know all elements in array must be in order

        # Result of 2.: Element at "partitioning index" has been pivoted to its correct position
        p_index = partition(array, low, high)
        # note we need to define this function `partition`, see below

        # 3. Recursively sort the elements below the partitioning index...
        _quick_sort(array, low, p_index-1)
        # ...and, separately, above the partitioning index:
        _quick_sort(array, p_index+1, high)

    return array
```

```python
# Final part - the algorithm itself
def quick_sort(array):
    _quick_sort(array, 0, len(array)-1)
    # to start, low is index of first element in array; high is penultimate
    # it resembles the use of recurssion
    return array
```
